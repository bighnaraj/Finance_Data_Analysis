```python
import pandas as pd                                 #First Two lines are needed
pd.core.common.is_list_like = pd.api.types.is_list_like # to patch the error
import pandas_datareader.data as web
import datetime
from bokeh.plotting import figure,show,output_file
#data.DataReader?

start = datetime.datetime(2015, 11, 1)
end = datetime.datetime(2016, 3, 10)

df = web.DataReader('GOOG','iex',start,end)

hours_12 = 12*3600*1000


def inc_dec(closed,opened):
    if closed > opened:
        value = "increase"
    elif closed < opened:
        value = "decrease"
    else:
        value = "equal"
    return value

df["middle"] = (df.open+df.close)/2
df["status"] = [inc_dec(closed,opened) for (closed,opened) in zip(df.close,df.open)]
df["height"] = abs(df.open-df.close)

l=[]
for item in df.index:
    values = item.split("-")
    l.append(datetime.datetime(int(values[0]),int(values[1]),int(values[2])))
df["converted_date"] = l
print(df)
 

p = figure(x_axis_type='datetime',width=1000,height=300)
p.title.text = "CandleStick Chart"
p.grid.grid_line_alpha = 0.3

p.segment(df.converted_date,df.high,df.converted_date,df.low,color="black")
p.rect(df.converted_date[df.status == "increase"],df.middle[df.status == "increase"],hours_12,df.height[df.status == "increase"],fill_color='green',line_color='black')
p.rect(df.converted_date[df.status == "decrease"],df.middle[df.status == "decrease"],hours_12,df.height[df.status == "decrease"],fill_color='red',line_color='black')

output_file("CS.html")
show(p)
```
